```python
import pandas as pd
import spacy
import random
from collections import Counter #for counting
import seaborn as sns
import os
```

```python

nlp=spacy.load('en_core_web_sm')
```

```python
#reading input fiels
tweets=pd.read_csv(r'D:\Naresh_it_praksah_senapathi\august\29th\6. NLP PROJECTS\NLP TASKS\TASK - 58\justdoit_tweets_2018_09_07_2.csv')
```

```python
#sample tweet text
#for simplicity let us take sample of tweets
random.seed(800)
text=tweets.tweet_full_text[random.sample(range(1,100),10)]
text
```

```python
#Let us begin our NLP journey with Lingustic Annotation, which means marking each and every word with its linguistic type like if it's a NOUN, VERB and so on. This help us in giving grammatical labels to our Text Corpus. 
#The function nlp() takes only string so let us use str() to combine all our rows above into one long string.
text_combined=str(text)
```

```python
doc=nlp(text_combined)
```

```python
doc
```

```python
#tokenizatin
for token in doc:
    print(token)
```

```python
#Since we have already done the annotation, 
#Let us print our chunks with their Parts-of-speech tags.
for token in doc:
    print(token.text,token.pos_)
```

```python
#That's good, We've got a bunch of chunks and their respective POS tags. Perhaps, 
#we don't want to see everything but just NOUNs. Below is the code how we can print only the nouns in the text.
nouns=list(doc.noun_chunks)
nouns
```

```python
#Sometimes, we might need to tokenization based on sentences. 
#Let's say we've got Chat Transcript from Customer Service and in that case we need to tokenize our transcript based on sentences.
list(doc.sents)
```

```python
'''Named Entity Recognition (NER)
NER is the process of extracting Named Entities like Person, Organization, Location and other such infromation from our Text Corpus. 
spaCy also has an object displacy that lets us visualize our text with NER. 
We can display Named Entities using the following code:'''
for ent in doc.ents:
    print(ent.text,ent.label_)
```

spaCy also allows to visualize Named Entities along woith the Text Labels.

```python
spacy.displacy.render(doc,style='ent',jupyter=True)
```

Lemmatization
Lemmetiztion is the process of retrieving the root word of the current word.
Lemmatization is an essential process in NLP to bring different variants of a single word to one root word.

```python
for token in doc:
    print(token.text,token.lemma_)
```

```python
#As you can see in the above output, words like aligning and values have been converted to their root words align and value.
```
